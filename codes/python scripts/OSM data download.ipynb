```python
import osmnx as ox
import geopandas as gpd
import pandas as pd
```

```python
place = ["Tbilisi, Georgia", 
         "Batumi, Georgia", 
         "Kutaisi, Georgia", 
         "Zugdidi, Georgia", 
         "Gori, Georgia", 
         "Poti, Georgia",
         "Sokhumi, Georgia",
         "Akhaltsikhe, Georgia"
        ]
```

```python
tags = {"amenity": ["cafe", "bar", "restaurant", "fast_food", "nightclub"]}
gdf = ox.features_from_place(place, tags)
gdf
```

```python
gdf.reset_index([0, 1])
new_gdf = gdf.reset_index(['element_type','osmid'])
new_gdf
```

```python
# Filter the GeoDataFrame to include only nodes, aka points
nodes_gdf = new_gdf[new_gdf['geometry'].type == 'Point']
nodes_gdf
```
